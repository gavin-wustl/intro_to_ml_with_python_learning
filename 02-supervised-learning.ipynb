```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
prediction = knn.predict(X_test)
print("Prediction on test set: ", prediction)
```

```python
print("KNN Classifier prediction accuracy: {:.3f}".format(np.mean(prediction == y_test)))
print("Accuracy of test set prediction: {:.3f}".format(knn.score(X_test, y_test)))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for n_neighbor, ax in zip([1,3,9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbor).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=0.4)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbor))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
    ax.legend(loc=4)
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []
neighbor_settings = range(1,11)

for neighbor in neighbor_settings:
    clf = KNeighborsClassifier(n_neighbors=neighbor)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbor_settings, training_accuracy, label="training accuracy")
plt.plot(neighbor_settings, test_accuracy, label="testing accuracy")
plt.xlabel("number of neighbors")
plt.ylabel("accuracy")
plt.legend(loc=1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
```

```python
prediction = reg.predict(X_test)
print("3-Neighbor regressor prediction: \n", prediction)
print('-'*30)
print("Prediction accuracy: {:.2f}".format(reg.score(X_test, y_test)))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,4))
line = np.linspace(-3, 3, 1000).reshape(-1, 1)

for n_neighbor, ax in zip([1, 3, 9], axes):
    reg = KNeighborsRegressor(n_neighbors=n_neighbor)
    reg.fit(X_train, y_train)
    prediction = reg.predict(line)
    ax.plot(line, prediction)
    ax.plot(X_train, y_train, '^')
    ax.plot(X_test, y_test, 'v')
    ax.set_title("{} neighbor(s) \n training score: {:.2f} \n testing score: {:.2f}".format(n_neighbor,
                                                                                    reg.score(X_train, y_train),
                                                                                    reg.score(X_test, y_test)))
```

```python
from sklearn.linear_model import LinearRegression
X, y = mglearn.datasets.make_wave(n_samples=60)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
```

```python
print("linear regression (lr) coefficient: ", lr.coef_)
print("linear regression (lr) intercept: ", lr.intercept_)
```

```python
print("training set score: {:.2f}".format(lr.score(X_train, y_train)))
print("test set score: {:.2f}".format(lr.score(X_test, y_test)))
```
