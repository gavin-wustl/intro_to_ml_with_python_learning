```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print('Keys of the iris dataset: \n',iris_dataset.keys())
```

```python
print(iris_dataset['DESCR'][:200], '\n')
print('-'*30)
print('Target Names: \n',iris_dataset['target_names'])
print('-'*30)
print('Feature Names: \n', iris_dataset['feature_names'])
print('-'*30)
print('Iris Data Shape: \n', iris_dataset['data'].shape)
print('-'*30)
print('First 5 Rows: \n', iris_dataset['data'][:5])
print('-'*30)
print('Type of Target: \n', type(iris_dataset['target']))
print('-'*30)
print('Shape of Target: \n', iris_dataset['target'].shape)
print('-'*30)
print('First five target: \n', iris_dataset['target'][:5])
print('-'*30)
print('File Name: \n', iris_dataset['filename'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
print('Training Set:')
print('X shape: ', X_train.shape)
print('y shape: ', y_train.shape)
print('-'*30)
print("Test Set:")
print('X shape: ', X_test.shape)
print('y shape: ', y_test.shape)
```

```python
feature_names = iris_dataset['feature_names']
iris_df_X = pd.DataFrame(X_train, columns=feature_names)
iris_df_X.head()
```

```python
pd.plotting.scatter_matrix(iris_df_X, c=y_train, figsize=(16,12), marker='o', hist_kwds={'bins':20}, s=60, alpha=0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5,2.9,1,0.2]])
X_new.shape
```

```python
prediction = knn.predict(X_new)
print('Prediction for X_new is: ', prediction)
print('Predicted species is: ', iris_dataset['target_names'][prediction])
```

```python
y_pred = knn.predict(X_test)
print('KNN Prediction: ',y_pred)
print(y_test)
```

```python
print('Prediction accuracy: {:.2f}'.format(np.mean(y_pred == y_test)))
print('Prediction accuracy using knn.score method: {:.2f}'.format(knn.score(X_test, y_test)))
```
